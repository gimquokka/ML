## Show in case of one perceptron nn!

```
# Initialize parameters
np.random.seed(42)
weights = np.random.rand(2, 1) 
lr = 0.5
bias = np.random.rand(1)

for epoch in range(1):
    inputs = feature_set

    # feedforward step 1
    XW = np.dot(feature_set,weights) + bias

    # feedforward step 2
    z = sigmoid(XW)

    # backpropagation step 1
    error_out = ((1 / 2) * (np.power((z - labels), 2)))
    print(error_out.sum())

    error = z - labels

    # backpropagation step 2
    dcost_dpred = error
    dpred_dz = sigmoid_der(z) 

    z_delta = dcost_dpred * dpred_dz

    inputs = feature_set.T
    weights -= lr * np.dot(inputs, z_delta)

    for num in z_delta:
        bias -= lr * num

# Show the output by the above training process
# feedforward step 1
XW = np.dot(feature_set,weights) + bias
# feedforward step 2
z = sigmoid(XW)

plt.figure(figsize=(10,7))
trained_labels = z
plt.scatter(feature_set[:,0], feature_set[:,1], c=z, cmap=plt.cm.winter)
```
